```python
import os
import glob
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import ResNet50V2
from tensorflow.keras.applications.resnet_v2 import preprocess_input
from tensorflow.keras import layers, models
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, jaccard_score
from scipy.optimize import linear_sum_assignment
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Configuration
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
N_CLASSES = 4  # Change/No-Change classes or Semantic classes (Tree, Building, etc.)
RANDOM_SEED = 42
```

```python
# --- UPDATE THESE PATHS ---
# Paths to your test set images (Time 1 and Time 2)
# Ensure filenames match or sort identically so T1_image_01 corresponds to T2_image_01
T1_DIR = 'SECOND_train_set/im1'  
T2_DIR = 'SECOND_train_set/im2'
LABEL_DIR = 'SECOND_total_test/label' # If available, for evaluation

def load_data_paths(t1_dir, t2_dir, label_dir=None):
    t1_paths = sorted(glob.glob(os.path.join(t1_dir, "*")))
    t2_paths = sorted(glob.glob(os.path.join(t2_dir, "*")))
    
    # Sanity check
    if len(t1_paths) != len(t2_paths):
        print(f"Warning: T1 has {len(t1_paths)} images, T2 has {len(t2_paths)}.")
        min_len = min(len(t1_paths), len(t2_paths))
        t1_paths = t1_paths[:min_len]
        t2_paths = t2_paths[:min_len]
    
    labels = None
    if label_dir:
        # Assuming labels are images/masks. If csv, adjust accordingly.
        label_paths = sorted(glob.glob(os.path.join(label_dir, "*")))
        # Placeholder for loading actual labels - adapting to your specific label format
        # Here we assume file-based labels for demonstration
        labels = np.array([0] * len(t1_paths)) # REPLACE with actual label loading logic
        
    print(f"Loaded {len(t1_paths)} image pairs.")
    return t1_paths, t2_paths, labels

# Execute Load
t1_paths, t2_paths, y_true = load_data_paths(T1_DIR, T2_DIR, LABEL_DIR)

# NOTE: Since we are doing unsupervised clustering, if you don't have labels loaded 
# correctly yet, we will generate dummy labels just so the code runs, 
# but the accuracy score will be meaningless until you plug in real labels.
if y_true is None or len(np.unique(y_true)) < 2:
    print("⚠️ No Ground Truth found. Using dummy labels for code verification.")
    y_true = np.random.randint(0, N_CLASSES, size=len(t1_paths))
```

```python
def build_siamese_extractor():
    # Base CNN
    base_model = ResNet50V2(weights='imagenet', include_top=False, input_shape=IMG_SIZE + (3,))
    base_model.trainable = False
    
    # Inputs
    img_input = tf.keras.Input(shape=IMG_SIZE + (3,))
    
    # Forward Pass
    x = preprocess_input(img_input)
    x = base_model(x, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    
    # Feature Vector
    return models.Model(inputs=img_input, outputs=x)

model = build_siamese_extractor()
print("Siamese Feature Extractor built.")
```

```python
def extract_siamese_features(model, t1_paths, t2_paths):
    features_list = []
    
    print("Extracting features from pairs...")
    for i in range(0, len(t1_paths), BATCH_SIZE):
        # Batch paths
        batch_t1 = t1_paths[i:i+BATCH_SIZE]
        batch_t2 = t2_paths[i:i+BATCH_SIZE]
        
        # Load Images
        imgs_t1 = np.array([tf.keras.utils.img_to_array(tf.keras.utils.load_img(p, target_size=IMG_SIZE)) for p in batch_t1])
        imgs_t2 = np.array([tf.keras.utils.img_to_array(tf.keras.utils.load_img(p, target_size=IMG_SIZE)) for p in batch_t2])
        
        # Extract Raw Features
        feat_t1 = model.predict(imgs_t1, verbose=0)
        feat_t2 = model.predict(imgs_t2, verbose=0)
        
        # --- STRATEGY: Difference Embedding ---
        # The paper emphasizes that change detection requires knowing "what" (semantic)
        # and "how much" (difference).
        # We concatenate: [T2_Semantic_State, Difference_Magnitude]
        diff_feat = np.abs(feat_t1 - feat_t2)
        
        # You can try concatenating [feat_t1, feat_t2, diff_feat] for full context
        # But often [feat_t2, diff_feat] is sufficient and lower dimension.
        combined = np.concatenate([feat_t2, diff_feat], axis=1)
        
        features_list.append(combined)
        
        if i % 100 == 0:
            print(f"Processed {i} pairs...")
            
    return np.vstack(features_list)

X_features = extract_siamese_features(model, t1_paths, t2_paths)
print(f"Feature Matrix Shape: {X_features.shape}")
```

```python
# Optional: PCA if dimensions are too high (ResNet out is 2048, so Combined is 4096)
print("Reducing dimensions with PCA...")
pca = PCA(n_components=0.95) # Keep 95% variance
X_pca = pca.fit_transform(X_features)
print(f"Dimensions reduced: {X_features.shape[1]} -> {X_pca.shape[1]}")

print("Running K-Means Clustering...")
kmeans = KMeans(n_clusters=N_CLASSES, random_state=RANDOM_SEED, n_init=10)
cluster_preds = kmeans.fit_predict(X_pca)
print("Clustering Complete.")
```

```python
# Hungarian Matching to map arbitrary cluster IDs to real class labels
def get_best_mapping(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred, labels=list(range(N_CLASSES)))
    row_ind, col_ind = linear_sum_assignment(cm.max() - cm)
    
    new_preds = np.zeros_like(y_pred)
    for r, c in zip(row_ind, col_ind):
        new_preds[y_pred == c] = r
    return new_preds

y_mapped = get_best_mapping(y_true, cluster_preds)

# Calculate Metrics
acc = accuracy_score(y_true, y_mapped)
# Estimate mIoU (Mean Intersection over Union) from the confusion matrix
# mIoU = TP / (TP + FP + FN) averaged over classes
cm = confusion_matrix(y_true, y_mapped)
iou_per_class = []
for i in range(N_CLASSES):
    tp = cm[i, i]
    fp = cm[:, i].sum() - tp
    fn = cm[i, :].sum() - tp
    denominator = tp + fp + fn
    if denominator == 0:
        iou = 0
    else:
        iou = tp / denominator
    iou_per_class.append(iou)

mIoU = np.mean(iou_per_class) * 100 # Convert to percentage

print(f"Accuracy: {acc:.4f}")
print(f"Estimated mIoU: {mIoU:.2f}%")
print("\nClassification Report:")
print(classification_report(y_true, y_mapped))
```
